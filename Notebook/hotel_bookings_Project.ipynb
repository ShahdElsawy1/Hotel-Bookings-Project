Detecting outliers(IQR)

```python
cols = ["adr", "lead_time"]
for col in cols:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers found")
```

#**Data Cleaning**

Handling Missing Values

```python
df[['company','agent']]=df[['company','agent']].fillna('None')
df['country']=df['country'].fillna('Unknown')
df['children']=df['children'].fillna(df['children'].median(),inplace=True)
```

Removing Duplicates

```python
df = df.drop_duplicates()
```

Fixing Data Types

```python
df.dtypes
```

```python
df['hotel'].value_counts()
```

```python
df['country'].value_counts()
```

```python
df["children"] = pd.to_numeric(df["children"])
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"])
```

```python
categorical_cols = [
    "hotel",
    "arrival_date_month",
    "meal",
    "country",
    "market_segment",
    "distribution_channel",
    "reserved_room_type",
    "assigned_room_type",
    "deposit_type",
    "agent",
    "company",
    "customer_type",
    "reservation_status"
] #A list of the columns that needs to be converted to category type

df[categorical_cols] = df[categorical_cols].astype("category")
```

#**Feature Engineering & Preprocessing**

New features

```python
import numpy as np
df['total_guests'] = df['adults'] + df['children'] + df['babies']
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df["is_family"] = np.where((df["children"] > 0) | (df["babies"] > 0), "Yes", "No")
```

Encoding Categorical Variables:

```python
from sklearn.preprocessing import OneHotEncoder
df_encoded = pd.get_dummies(df, columns=["meal", "market_segment"], drop_first=True)
df_encoded.head()
```

```python
freq_encoding=df['country'].value_counts(normalize=True)
df['country Freq']=df['country'].map(freq_encoding)
df['country Freq'].head()
```

```python
top10=df['country'].value_counts().nlargest(10).index
df['country Top10']=np.where(df['country'].isin(top10),df['country'],'Other')
df['country Top10'].value_counts()
```

```python
df = df.drop(columns=["reservation_status", "reservation_status_date"])
```

Splitting Dataset

```python
from sklearn.model_selection import train_test_split
x=df.drop('is_canceled',axis=1)
y=df['is_canceled']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```
